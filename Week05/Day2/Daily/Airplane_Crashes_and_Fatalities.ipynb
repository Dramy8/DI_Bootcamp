```python
# copy of the original dataset
df_clean = df.copy()
```

```python
# Drop column with too many missing values and low analytical value
df_clean = df_clean.drop(columns=["Flight #"])
```

```python
# Fill categorical/text missing values
cols_unknown = [
    "Time",
    "Location",
    "Operator",
    "Route",
    "AC Type",
    "Registration",
    "cn/ln"  # Corrected from "cn/In"
]

for col in cols_unknown:
    df_clean[col] = df_clean[col].fillna("Unknown")
```

```python
# Fill missing summaries
df_clean["Summary"] = df_clean["Summary"].fillna("No summary available")
```

```python
# Drop rows where key numerical columns are missing
df_clean = df_clean.dropna(subset=["Aboard", "Fatalities"])
```

```python
df_clean.isnull().sum()
```

```python
df_clean
```

```python
df_clean.columns = df_clean.columns.str.strip()
```

```python
# convert in date format
df_clean = df_clean.copy()

df_clean["Date"] = pd.to_datetime(df_clean["Date"], errors="coerce")
```

```python
# useful date columns
df_clean["Year"] = df_clean["Date"].dt.year
df_clean["Month"] = df_clean["Date"].dt.month
df_clean["Decade"] = (df_clean["Year"] // 10) * 10
```

```python
# convert num columns in numerical value
num_cols = [
    "Aboard",
    "Aboard Passangers",  # Corrected from 'Aboard Passengers'
    "Aboard Crew",
    "Fatalities",
    "Fatalities Passangers",  # Corrected from 'Fatalities Passengers'
    "Fatalities Crew",
    "Ground"
]

for col in num_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")
```

```python
# checking conversion
df_clean[num_cols].info()
```

```python
# creation of new column Survivor , analysis

df_clean["Survivors"] = df_clean["Aboard"] - df_clean["Fatalities"]
```

```python
df_clean["Survival Rate"] = df_clean["Survivors"] / df_clean["Aboard"] * 100
```

```python
# fatality rate

df_clean["Fatality Rate"] = df_clean["Fatalities"] / df_clean["Aboard"] * 100
```

```python
# separating cat and num columns
num_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df_clean.select_dtypes(include=["object"]).columns
```

Missing values were handled depending on the type and importance of each column. The `Flight #` column was dropped because it had a high percentage of missing values and limited analytical value. Categorical missing values such as `Route`, `Operator`, and `AC Type` were replaced with `"Unknown"`. For key numerical columns such as `Aboard` and `Fatalities`, rows with missing values were removed because these variables are central to the analysis.

Hypothesis test

Comparing the average number of fatalities before 1980 and after 1980

```python
before_1980 = df_clean[df_clean["Year"] < 1980]["Fatalities"].dropna()
after_1980 = df_clean[df_clean["Year"] >= 1980]["Fatalities"].dropna()
```

```python
print("Before 1980:", before_1980.shape[0])
print("After 1980:", after_1980.shape[0])
```

```python
print(f"Mean fatalities before 1980: {before_1980.mean():.2f}")
print(f"Mean fatalities from 1980 onward: {after_1980.mean():.2f}")
```

```python
df_clean["Period"] = np.where(df_clean["Year"] < 1980, "Before 1980", "1980 and after")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_clean, x="Period", y="Fatalities")
plt.title("Fatalities Before and After 1980")
plt.xlabel("Period")
plt.ylabel("Number of Fatalities")
plt.show()


plt.figure(figsize=(8, 5))
sns.barplot(data=df_clean, x="Period", y="Fatalities", errorbar="ci")
plt.title("Average Fatalities Before and After 1980")
plt.xlabel("Period")
plt.ylabel("Average Fatalities")
plt.show()
```

```python
t_stat, p_value = stats.ttest_ind(before_1980, after_1980, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0: There is a statistically significant difference between the two periods.")
else:
    print("Fail to reject H0: There is no statistically significant difference between the two periods.")
```

Since the p-value is below 0.05, we reject the null hypothesis. This suggests that the average number of fatalities differs significantly between the two periods.
