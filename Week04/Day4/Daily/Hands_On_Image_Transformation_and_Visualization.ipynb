```python
# Load an image using PIL

from PIL import Image
import matplotlib.pyplot as plt

image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
  return rotate(image, 30, reshape=True) #complete it

img = rotate_image_30_degrees(original_image)
plt.imshow(img)
```

```python
from PIL.ImageOps import mirror, flip

def mirror_image(image):
      return mirror(image)

img1 = mirror_image(original_image)
plt.imshow(img1)
```

```python
def flip_image(image):
      return flip(image)

img2 = flip_image(original_image)
plt.imshow(img2)
```

```python
def zoom_image(image):
    width, height = image.size

    new_width = int(width*1.2)
    new_height = int(height*1.2)

    return image.resize((new_width, new_height))

img3 = zoom_image(original_image)
plt.imshow(img3)
```

```python
print(original_image.size)
print(img3.size)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(original_image)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(img3)
plt.title("Zoom 1.2x")

plt.show()
```
